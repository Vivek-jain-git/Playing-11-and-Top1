```python
def countteam():
    df1=pd.read_csv(r'C:\Users\Ryzen 5\Desktop\ipl-matches.csv')
    point=pd.read_csv(r'C:\Users\Ryzen 5\Desktop\ipl_points_table.csv',header=0).T
    point.columns=point.iloc[0,0:]
    point=point.iloc[1:]
    point=point.T
    point.rename(columns={'2020':'2020/21'},inplace=True)
    df1.replace({'Kings XI Punjab':'Punjab Kings','Rising Pune Supergiants':'Rising Pune Supergiant','Delhi Daredevils':'Delhi Capitals','Pune Warriors':'Pune Warriors India'},inplace=True)
    new_df=pd.DataFrame(columns=['Team','Team_Players','count','season','standings'])
    main_df=pd.DataFrame(columns=['Team','Team_Players','count','season','standings'])
    for a in df1['Season'].unique():
        df=df1[df1['Season']==a]
        for i in sorted(df['Team1'].unique()):
            df2=df[(df['Team1']==i) | (df['Team2']==i)]
            l1=[]
            for j in df2[df2['Team1']==i]['Team1Players']:
                l1.append(ast.literal_eval(j))
            l2=[]
            for k in df2[df2['Team2']==i]['Team2Players']:
                l2.append(ast.literal_eval(k))
            tl=l1+l2
            d={}
            for f in range(len(tl)):
                l=[]
                t=sorted(tl[f])
                for g in range(len(tl)):
                    if t==sorted(tl[g]):
                        l.append(g)
                d[tuple(t)]=len(l) 
            new_df=new_df.append({'Team':i,'Team_Players':sorted(d.items(),key=lambda x:x[1],reverse=True)[0][0],'count':sorted(d.items(),key=lambda x:x[1],reverse=True)[0][1],'season':a,'standings':point.loc[i,a]},ignore_index=True)
        main_df=main_df.append(new_df)
    final=pd.DataFrame(columns=['Team','Team_Players','count','season','standings'])
    for i in point.columns[::-1]:
        final=final.append(main_df.groupby('season').get_group(i).drop_duplicates(keep='first'))
    final['standings']=final['standings'].astype('int32')
    final['count']=final['count'].astype('int32')
    return np.corrcoef(final['count'],final['standings'])
a=countteam()        
a
```
